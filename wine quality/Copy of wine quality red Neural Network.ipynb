#Classifier

```python
from sklearn.neural_network import MLPClassifier
model=MLPClassifier()
```

```python
#Please see the following website for reference
#https://towardsdatascience.com/hyperparameter-tuning-the-random-forest-in-python-using-scikit-learn-28d2aa77dd74

import numpy as np

# Number of hidden layers for the neural network
hidden_layer_sizes = [(10,),(5,5),(4,4,4),(3,3,3,3),(2,2,2,2,2)]

# activation function for neural network
#x=(a1)x1+(a2)x2+(a3)x3+....(an)(xn) 
#identity: f(x)=x
# logistic: f(x)=
# tanh:
# relu:
activation = ['logistic','tanh','relu']

# Maximum number of levels in tree
solver = ['lbfgs','sgd','adam']

# Minimum number of samples required to split a node
alpha = [0.0005, 0.001,0.005]

# Method of selecting samples for training each tree
learning_rate = ['constant', 'invscaling','adaptive']

# Number of features to consider at every split
power_t = [0.25, 0.5,0.75]

# Maximum number of levels in tree
max_iter = [500,1000,1500,2000]

# Method of selecting samples for training each tree
tol = [1e-5,1e-4,1e-3]

# Minimum number of samples required to split a node
n_iter_no_change = [10, 50, 100]


# Create the random grid
random_grid = {'hidden_layer_sizes': hidden_layer_sizes,
               'activation': activation,
               'solver': solver,
               'alpha': alpha,
               'learning_rate': learning_rate,
               'power_t': power_t,
               'max_iter': max_iter,
               'tol': tol,
               'n_iter_no_change': n_iter_no_change}
print(random_grid)
```

```python
from sklearn.model_selection import RandomizedSearchCV

# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
model_random= RandomizedSearchCV(estimator=model,param_distributions= random_grid, n_iter=100, cv = 3, verbose=2, n_jobs=-1)
```

```python
model_random.fit(dataFeatures,dataPhenotypes)
model_random.best_params_
```

```python
from sklearn.model_selection import GridSearchCV

# Create the random grid
param_grid = {
 'activation': ['tanh','relu'],
 'alpha': [0.0005, 0.001,0.005],
 'hidden_layer_sizes': [(2, 2, 2, 2, 2),(5, 5),(10,)],
 'learning_rate': ['adaptive','invscaling'],
 'max_iter': [1000],
 'n_iter_no_change': [10,50],
 'power_t': [0.5,0.75],
 'solver': ['lbfgs','sgd','adam'],
 'tol': [0.0001,1e-05]}
print(param_grid)
```

```python
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
model_grid= GridSearchCV(estimator=model,param_grid= param_grid, cv = 3, verbose=2, n_jobs=-1)
```

```python
model_grid.fit(dataFeatures,dataPhenotypes)
model_grid.best_params_
```

```python
from sklearn.neural_network import MLPClassifier
model=MLPClassifier(activation='relu',alpha=0.001, hidden_layer_sizes=(10,), learning_rate='adaptive',max_iter=1000, n_iter_no_change=50, power_t=0.75, solver='lbfgs',tol=1e-05)


TrainedModel=model.fit(dataFeatures,dataPhenotypes)
```

# Real curve

```python
from sklearn.metrics import roc_curve, auc, precision_recall_curve
import matplotlib.pyplot as plt

probs = TrainedModel.predict_proba(dataFeatures_true)
fpr1, tpr1, thresholds1 = roc_curve(DataPhenotypes_true, probs[:,1])
prec1, recall1, thresholds1 = precision_recall_curve(DataPhenotypes_true, probs[:,1])

plt.plot(fpr1,tpr1,label="ROC")
plt.plot(recall1,prec1,label="PRC")

plt.xlabel('False Positive Rate/Recall')
plt.ylabel('True Positive Rate/Precision')

plt.legend()
plt.title('True')
plt.show()
plt.savefig("wine quality red Neural Network true.png")

print("PRC AUC:" + str(auc(recall1, prec1)))
print("ROC AUC:" + str(auc(fpr1, tpr1)))
```
