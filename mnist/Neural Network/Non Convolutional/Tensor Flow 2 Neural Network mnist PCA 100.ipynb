```python
import tensorflow as tf
from sklearn.preprocessing import LabelBinarizer
```

# **Data Setup**

---

```python
mnist=tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test= x_train/ 255.0, x_test /255.0
Y_train=LabelBinarizer().fit_transform(y_train)
Y_test=LabelBinarizer().fit_transform(y_test)
```

```python
a=len(x_train)
b=len(x_train[0])
c=len(x_train[0][0])
d=b*c

x_train_1=x_train.reshape([a,d])
```

# **PCA**

---

```python
from sklearn.decomposition import PCA
x_train_reduced = PCA(n_components=100).fit_transform(x_train_1)
```

# **Model Setup**

---

```python
from sklearn.neural_network import MLPClassifier
model=MLPClassifier(hidden_layer_sizes=(50,), activation='logistic',solver='sgd',max_iter=50, random_state=1)


TrainedModel=model.fit(x_train_reduced,Y_train)
```

```python
e=len(x_test)
f=len(x_test[0])
g=len(x_test[0][0])
h=f*g

x_test_1=x_test.reshape([e,h])
```

```python
x_test_reduced = PCA(n_components=100).fit_transform(x_test_1)
```

```python
from sklearn.metrics import roc_curve, auc, precision_recall_curve
import matplotlib.pyplot as plt

probs = TrainedModel.predict_proba(x_test_reduced)
fpr1, tpr1, thresholds1 = roc_curve(Y_test[:,0], probs[:,0])
prec1, recall1, thresholds1 = precision_recall_curve(Y_test[:,0], probs[:,0])

plt.plot(fpr1,tpr1,label="ROC")
plt.plot(recall1,prec1,label="PRC")

plt.xlabel('False Positive Rate/Recall')
plt.ylabel('True Positive Rate/Precision')

plt.legend()
plt.title('0')
plt.show()

print("PRC AUC:" + str(auc(recall1, prec1)))
print("ROC AUC:" + str(auc(fpr1, tpr1)))
```
